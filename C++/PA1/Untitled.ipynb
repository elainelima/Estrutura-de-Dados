```python
from abc import ABC, abstractmethod
class Professor(ABC):
    def __init__(self, nome : str, email : str):
        if not isinstance(nome, str):
            raise TypeError("O nome deve ser do tipo string.")
        if not isinstance(email, str):
            raise TypeError("O email deve ser do tipo string.")
        self.nome = nome
        self.email = email
    def get_nome(self) -> str:
        return self.nome
    def get_email(self) -> str:
        return self.email
    @abstractmethod
    def get_salario() -> float:
        pass

class ProfessorHorista(Professor):
    def __init__(self, nome, email, horasTrabalhadas, valorHora):
        super().__init__(nome, email)
        if not isinstance(horasTrabalhadas, (int, float)):
            raise TypeError("As horas trabalhadas devem ser do tipo número.")
        if not isinstance(valorHora, (int, float)):
            raise TypeError("O valor da hora deve ser do tipo número.")
        if horasTrabalhadas < 0:
            raise ValueError("O número de horas trabalhadas não pode ser negativo.")
        if valorHora < 10:
            raise ValorHoraInvalidoError()
        self.horasTrabalhadas = horasTrabalhadas
        self.valorHora = valorHora
    def get_salario(self):
        return self.horasTrabalhadas * self.valorHora
    
class ProfessorIntegral(Professor):
    def __init__(self, nome, email, salarioMensal):
        super().__init__(nome, email)
        if not isinstance(salarioMensal, (int, float)):
            raise TypeError("O salário deve ser do tipo número.")
        if salarioMensal <= 0:
            raise ValueError("O salário deve ser maior que zero.")
        self.salarioMensal = salarioMensal
    def get_salario(self):
        return self.salarioMensal
class Universidade():
    def __init__ (self, listaProfessor = []):
        self.listaProfessor = []
    def adiciona_professor(self, professor):
        if not isinstance(professor, Professor):
            raise TypeError("O objeto adicionado deve ser uma instância da classe Professor.")
        if professor in self.listaProfessor:
            print(f"Professor {professor.nome} já está na lista.")
            return
        self.listaProfessor.append(professor)
    def listar_professores(self):
        for professor in self.listaProfessor:
            print(f"Nome: {professor.nome}, Email: {professor.email}")
    def total_salario(self):
        totalSalario = 0
        for professor in self.listaProfessor:
            totalSalario = totalSalario + professor.get_salario()
        return totalSalario

        
```

```python
from abc import ABC, abstractmethod
import math
class NumeroSubdivisoesInvalidoError(Exception):
    #Excecao personalizada
    def __init__(self, n):
        super().__init__(f"O número de subdivisões deve ser maior ou igual a 1. N = {n}")
class Integral():
    def __init__(self):
        pass
    @abstractmethod
    def resultado(self, x):
        pass
    def integral(self, a, b, n):
        if not isinstance(a, (int, float)):
            raise TypeError("O limite inferior a deve ser do tipo número.")
        if not isinstance(b, (int, float)):
            raise TypeError("O limite superior b deve ser do tipo número.")
        if not isinstance(n, int):
            raise TypeError("O número de subdivisões n deve ser do tipo número inteiro.")
        if a >= b:
            raise ValueError("O limite inferior a deve ser menor que o limite superior b.")
        if n < 1:
            raise NumeroSubdivisoesInvalidoError(n)
        h = (b - a) / n
        total = 0.5 * (self.resultado(a) + self.resultado(b))
        for i in range(1, n):
            total += self.resultado(a + i * h)
        return total*h
class Quadratica(Integral):
    def __init__(self,a1, a2, a3):
        self.a1 = a1
        self.a2 = a2
        self.a3 = a3
    def resultado(self, x):
        return self.a1*(x**2) + self.a2*x + self.a3
        
class Linear(Integral):
    def __init__(self, a1, a2):
        self.a1 = a1
        self.a2 = a2
    def resultado(self, x):
        return self.a1 * x + self.a2

class Seno(Integral):
    def resultado(self, x):
        if x == 0:
            return 0  
        return math.sin(x)/ x
        
```
